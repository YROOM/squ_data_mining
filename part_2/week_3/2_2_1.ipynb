```python
import numpy as np
x = np.arange(10)
print(x)
```

```python
y = x[2:6] #切片索引  只是原数组的视图
print(y)
y[0]=8888
```

```python
x= np.arange(10,1,-1)
print(x)
```

```python
list = [2,2,1,6]
y = x[list]
print(y)
y[0]=88888
print(x)
print(y)
```
